```python
import json
import os
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
from getdata import *
from google.colab import drive
drive.mount('/content/drive')
# Ruta de la carpeta "Results_30_10" (por dentro debe contener más directorios con los resultados)
path_selected_stats = "/content/drive/MyDrive/ResultadosArquitectura/results_30_10/results_V2"
# Ruta al csv generado por las simulaciones
csv_path = "/content/drive/MyDrive/ResultadosArquitectura/results_30_10/results_V2.csv"
stats_all = extract_simulation_data(path_selected_stats, csv_path)
```

```python
# === Conversión a números flotantes ===
for s in stats_all.values():
    for k, v in s.items():
        s[k] = float(v)

# Define the parameters to plot
parameters_to_plot = ["Accuracy_bp", "bp_btb_hitrate", "Accuracy_dcache", "cpi", "ipc", "energy", "edp"]
stat_names = list(stats_all.keys())

# Calculate the derived metrics for each statistic
calculated_metrics = {}
for stat_name, stat_data in stats_all.items():
    calculated_metrics[stat_name] = {
        "Accuracy_bp": (1 - stat_data["branchPred_btb_mispredict"] / stat_data["branchPred_total_btb_lookups"]),
        "bp_btb_hitrate": stat_data["branchPred_btb_hitrate"],
        "Accuracy_dcache": stat_data["dcache_demandHits_total"] / (stat_data["dcache_demandHits_total"] + stat_data["dcache_demandMisses"]),
        "cpi": stat_data["cpi"],
        "ipc": 1 / stat_data["cpi"],
        "energy": stat_data["energy"],
        "edp": stat_data["edp"]
    }

# Determine the number of rows and columns for the subplots
n_parameters = len(parameters_to_plot)
n_cols = 2 # You can adjust the number of columns
n_rows = (n_parameters + n_cols - 1) // n_cols

# Create the figure and subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 5))
axes = axes.flatten() # Flatten the 2D array of axes for easy iteration

# Get a color map
colors = cm.viridis(np.linspace(0, 1, len(stat_names)))

# Plot each parameter in a separate subplot
for i, parameter in enumerate(parameters_to_plot):
    ax = axes[i]
    parameter_values = [calculated_metrics[stat_name][parameter] for stat_name in stat_names]
    bars = ax.bar(range(len(parameter_values)), parameter_values, color=colors)

    # Add values to the bars
    def autolabel(rects):
        """Attach a text label above each bar in *rects*, displaying its height."""
        for rect in rects:
            height = rect.get_height()
            # ax.annotate('%.2f' % height,
            #             xy=(rect.get_x() + rect.get_width() / 2, height),
            #             xytext=(0, 3),  # 3 points vertical offset
            #             textcoords="offset points",
            #             ha='center', va='bottom')

    autolabel(bars)

    ax.set_ylabel(parameter)
    ax.set_title(f"{parameter} across Simulations")
    ax.tick_params(axis='x', rotation=45)
    ax.grid(axis='y', linestyle='--', alpha=0.7)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

## Instruction Profiling

```python
inst_branch = stats_all["Stat0"]["inst_branch"]
inst_load = stats_all["Stat0"]["inst_load"]
inst_store = stats_all["Stat0"]["inst_store"]
inst_ALU = stats_all["Stat0"]["inst_ALU"]

total_inst = inst_branch + inst_load + inst_store + inst_ALU
branch_prop = inst_branch / total_inst
load_prop = inst_load / total_inst
store_prop = inst_store / total_inst
ALU_prop = inst_ALU / total_inst


# Plot the distribution of the instruction in 1 single stacked bar
fig, ax = plt.subplots()

ax.bar("Instruction Distribution", branch_prop, label="Branch", color="blue")
ax.bar("Instruction Distribution", load_prop, bottom=branch_prop, label="Load", color="green")
ax.bar("Instruction Distribution", store_prop, bottom=branch_prop + load_prop, label="Store", color="red")
ax.bar("Instruction Distribution", ALU_prop, bottom=branch_prop + load_prop + store_prop, label="ALU", color="orange")

ax.set_ylabel("Proportion of Instructions")
ax.set_title("Profiling")
ax.legend()
plt.show()
```

```python
best_performance = sorted(calculated_metrics.items(), key=lambda x: x[1]['ipc'], reverse=True)[:3]
best_energy = sorted(calculated_metrics.items(), key=lambda x: x[1]['energy'], reverse=False)[:3]
best_edp = sorted(calculated_metrics.items(), key=lambda x: x[1]['edp'], reverse=False)[:3]

#plot, in x axis the name of the stat, y the value, make subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Function to add labels to bars
def autolabel(rects, ax):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('%.3f' % height,
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Plot Best 3 in Performance
bars_performance = axes[0].bar([x[0] for x in best_performance], [x[1]['ipc'] for x in best_performance], color = "green", width=0.3)
axes[0].set_ylabel("IPC")
axes[0].set_title("Best 3 in Performance")
autolabel(bars_performance, axes[0])


# Plot Best 3 in Energy
bars_energy = axes[1].bar([x[0] for x in best_energy], [x[1]['energy'] for x in best_energy], color = "orange", width=0.3)
axes[1].set_ylabel("Energy")
axes[1].set_title("Best 3 in Energy")
autolabel(bars_energy, axes[1])

# Plot Best 3 in EDP
bars_edp = axes[2].bar([x[0] for x in best_edp], [x[1]['edp'] for x in best_edp], width=0.3)
axes[2].set_ylabel("EDP")
axes[2].set_title("Best 3 in EDP")
autolabel(bars_edp, axes[2])

plt.tight_layout()
plt.show()
```
